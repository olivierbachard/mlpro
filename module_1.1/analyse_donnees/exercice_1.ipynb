# Exercice 1: Health Expectancy

```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

df = sns.load_dataset("healthexp")
df.head()
```

### Analysez la variable "Country"

```python
pd.Series(df["Country"]).value_counts()
```

### Analysez la variable "Spending_USD".

- Réalisez les 2 graphiques utiles pour cette analyse

```python
pd.Series(df["Spending_USD"]).describe()
```

Depuis 1970, les pays dépensent en moyenne 2789$ par habitant par an en frais de santé.

```python
# bar chart des dépenses moyennes par pays
# df.groupby("Country")["Spending_USD"].mean().plot(kind="bar")
sns.catplot(data=df, x="Country", y="Spending_USD", kind="bar")
# histogramme des dépenses moyennes par pays
sns.displot(data=df, x="Spending_USD", hue="Country")
# boite à moustache des dépenses par pays
sns.catplot(data=df, x="Spending_USD", y="Country", kind="box")
```

```python
# Life expectancy depending on spending
sns.scatterplot(df, x="Life_Expectancy", y="Spending_USD")
plt.show()
```

### Faites de même pour la variable "Life_Expectancy"

### Observez l'évolution de la durée de vie à travers le temps.

```python
sns.catplot(df, y="Life_Expectancy", x="Year", kind="bar")
plt.show()
```

### Observez l'évolution des dépenses en santé de l'Allemagne et de la Grande Bretagne à travers le temps.

```python
df2 = df.loc[df['Country'].isin(['Germany', 'Great Britain'])]
#print(df)
df2.head()
sns.catplot(df2, x="Year", y="Spending_USD")
```

### Observez l'histograme, la boxplot, et les statistiques de "Life_Expectancy" propre à chaque pays.

```python
# histogramme de l'espérance de vie 
sns.displot(data=df, y="Life_Expectancy", x="Country")
sns.catplot(data=df, x="Life_Expectancy", hue="Country", kind="box")
df.groupby("Country")["Life_Expectancy"].describe()
```

### Observez la relation entre la durée de vie et les dépenses dans la santé, aux États-Unis, de 2000 à 2020

```python
df3 = df.loc[(df["Year"] >= 2000) & (df["Year"] <= 2020) & (df["Country"] == 'USA')]
sns.catplot(df3, y="Spending_USD", x="Year", kind="bar")
sns.catplot(df3, y="Spending_USD", x="Life_Expectancy")
plt.show()
```
